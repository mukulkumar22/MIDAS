importing required libraries

```python
import jsonlines
import tweepy
import pandas as pd
import numpy as np
```

Consumer key and access token to get tweets using twitter API (APIs are ommited due to privacy issues)

```python
consumer_key = ""
consumer_secret=""
access_token=""
access_secret = ""
```

Authorising access to the twitter API

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
```

Get all the tweets as Status object using tweepy.Cursor and setting tweet_mode as "extended" to get the informaton about media files(images etc) if any.

```python
a = tweepy.Cursor(api.user_timeline, id="midasIIITD", tweet_mode='extended').items()
```

Write all the tweets extracted to a jsonline file names "output.jsonl"

```python
with jsonlines.open('output.jsonl', mode='w') as writer:
    for i in a:
        writer.write(i._json)
        
```

# Reading tweets from jsonline file
After dumping of tweets into jsonline file is done, let's read that file and extract the required information

First, let's create an empty dictionary to keep record of the required information of every tweet, then we will convert this to a dataframe using pandas to represent the data in a better way

```python
Table = {"Text" : [],"Created at" :[],"Number of Favorites" :[],"Number of retweets":[],"Number of images" :[]}
```

Reading data
<br>
If a tweet has any type of media(image, video etc.) then it contains attribute called __media__ under the attribute **entities**. An attribute named __"extended_entities"__ is also present when there are more than 1 media files.
<br>
A retweet has an attribute named __retweeted_status__ and the media is under __entities__ or __extended_entities__.
So, we will check for these requirements and return the length of the list of media if present otherwise None.

```python
with jsonlines.open('output.jsonl') as reader:
    for obj in reader:
        if 'retweeted_status' in obj and "media" in obj["retweeted_status"]["entities"]:
            if "extended_entities" in obj["retweeted_status"]:
                count = len(obj["retweeted_status"]["extended_entities"])
            else:
                count = 1
        elif "media" in obj["entities"]:
            if "extended_entities" in obj:
                count = len(obj["extended_entities"]["media"])
            else:
                count  = 1
            
        else:
            count = "None"
        
        """After counting images, let's add the other info to our Table.
           Read the required info and append it to the table under the suitable column"""
        
        Table["Number of images"].append(count)
        Table["Text"].append(obj["full_text"])
        Table["Created at"].append(obj["created_at"])
        Table["Number of Favorites"].append(obj["favorite_count"])
        Table["Number of retweets"].append(obj["retweet_count"])
            
    
```

## Creating dataframe from Table

```python
df = pd.DataFrame(data = Table)   #converted the dictionary to a dataframe using pandas library
pd.set_option('display.max_colwidth', -1)  #setting the display property to show the full(non-truncated) text of tweet 
df.index = np.arange(1, len(df) + 1) #indexing the dataframe from 1 instead of 0
```
